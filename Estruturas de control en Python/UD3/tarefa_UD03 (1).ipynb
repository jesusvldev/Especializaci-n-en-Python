# Exercicio 12: Usar match para verificar o tipo de froita
Instrucións: Define unha variable 'froita' cun nome de froita. Usa a sentenza match para imprimir unha mensaxe diferente dependendo da froita. Se a froita é "mazá", imprime "É unha mazá vermella.". Se é "banana", imprime "É unha banana amarela.". Para calquera outra froita, imprime "É outra froita.".

 -Saída esperada (se froita = "mazá"): É unha mazá vermella.

 -Saída esperada (se froita = "banana"): É unha banana amarela.

 -Saída esperada (se froita = "laranxa"): É outra froita.


```python
froita = "banana"

match froita:
    case "mazá":
        print("É unha mazá vermella.")
    case "banana":
        print("É unha banana amarela.")
    case _:
        print("É outra froita.")
```

# Exercicio 13: Usar match para verificar a puntuación dun exame
Instrucións: Define unha variable 'puntuacion' cun valor enteiro entre 0 e 100. Usa a sentenza match para imprimir unha mensaxe baseada na puntuación:

- Se a puntuación é 100, imprime "Puntuación perfecta!"

- Se a puntuación está entre 90 e 99, imprime "Excelente."

- Se a puntuación está entre 70 e 89, imprime "Notable."

- Se a puntuación está entre 50 e 69, imprime "Aprobado."

- Para calquera outra puntuación (menor de 50), imprime "Suspenso."

 Nota: Para rangos, podes usar gardas con if. Ex: case p if 90 <= p <= 99:

 Saída esperada (se puntuacion = 95): Excelente.

 Saída esperada (se puntuacion = 60): Aprobado.

 Saída esperada (se puntuacion = 40): Suspenso.

```python
p = 200

match p:
  case p if p == 100:
    print('Excelente')
  case p if 90 <= p <= 99:
    print('Sobresaliente')
  case _:
    print('no se que metiste')
```

# Exercicio 14: Usar match co caso comodín (_)
Instrucións: Define unha variable 'comando' cunha cadea de texto. Usa a sentenza match para simular a execución de comandos:

 - Se o comando é "iniciar", imprime "Iniciando o sistema."

 - Se o comando é "deter", imprime "Detendo o sistema."

 - Para calquera outro comando, usa o caso comodín (_) e imprime "Comando descoñecido: [comando]".

 Saída esperada (se comando = "iniciar"): Iniciando o sistema.

 Saída esperada (se comando = "deter"): Detendo o sistema.

 Saída esperada (se comando = "reiniciar"): Comando descoñecido: reiniciar

```python
comando = "patata"

match comando:
    case "iniciar":
        print("Iniciando o sistema.")
    case "deter":
        print("Detendo o sistema.")
    case _:
        print(f"Comando descoñecido: {comando}")

```

# Exercicio 15: Usar match para verificar tipos de datos ou patróns simples
Instrucións: Define unha variable 'dato' con diferentes tipos de datos (pode ser un número, unha cadea, unha lista baleira, unha lista con elementos, etc.). Usa a sentenza match para imprimir unha mensaxe diferente dependendo do tipo de dato ou patrón:

 - Se o dato é un número enteiro ou decimal, imprime "É un número.".

 - Se o dato é unha cadea de texto, imprime "É unha cadea de texto.".

 - Se o dato é unha lista baleira, imprime "É unha lista baleira.".

 - Se o dato é unha lista con polo menos un elemento, imprime "É unha lista con elementos.".

 - Para calquera outro tipo de dato, usa o caso comodín (_) e imprime "Tipo de dato descoñecido.".

 Saída esperada (se dato = 10): É un número.

 Saída esperada (se dato = "ola"): É unha cadea de texto.

 Saída esperada (se dato = []): É unha lista baleira.

 Saída esperada (se dato = [1, 2]): É unha lista con elementos.

 Saída esperada (se dato = (1, 2)): Tipo de dato descoñecido.

```python
dato = [1, 2]  # proba cambiando este valor

match dato:
    case int() | float():   # número enteiro ou decimal
        print("É un número.")
    case str():             # cadea de texto
        print("É unha cadea de texto.")
    case []:                # lista baleira
        print("É unha lista baleira.")
    case [*_]:              # lista con polo menos un elemento
        print("É unha lista con elementos.")
    case _:                 # comodín (calquera outro tipo)
        print("Tipo de dato descoñecido.")
```
