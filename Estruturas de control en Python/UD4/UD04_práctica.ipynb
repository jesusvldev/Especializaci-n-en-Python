# Exercicio 9:

Implementa unha función chamada sumar_lista que reciba unha lista de números e use un bucle for para calcular e devolver a suma dos elementos.

Ejemplo de saída:

*Proba de sumar_lista con [10, 20, 30, 40]:*

*A suma dos elementos da lista é: 100*


```python
#Escribe aquí o teu código
def sumar_lista(lista):
    suma = 0
    for numero in lista:
        suma += numero
    return suma


# Exemplo de uso
numeros = [10, 20, 30, 40]
print(f"Proba de sumar_lista con {numeros}:")
print("A suma dos elementos da lista é:", sumar_lista(numeros))

```

# Exercicio 12:
Define unha función chamada contar_elementos_unicos que reciba unha lista e use conxuntos para contar e devolver o número de elementos únicos na lista. Inclúe unha descrición en galego e un exemplo.

Exemplo de saída:

*A lista é: [1, 2, 2, 3, 4, 4, 5]*

*O número de elementos únicos na lista é: 5*

```python
# Escribe aquí o teu código
def contar_elementos_unicos(lista):

    return len(set(lista))


# Exemplo de uso
lista = [1, 2, 2, 3, 4, 4, 5]
print("A lista é:", lista)
print("O número de elementos únicos na lista é:", contar_elementos_unicos(lista))
```

# Exercicio 13:
Implementa unha función chamada validar_contrasinal que reciba unha cadea de texto e use múltiples condicións (if, elif, else) anidadas ou encadeadas para verificar se a contrasinal cumpre certos criterios (por exemplo, lonxitude mínima, contén maiúsculas, minúsculas e números). Devolve True se é válida e False en caso contrario. Inclúe unha descrición en galego e un exemplo.

Exemplo de saída:

*Proba de validar_contrasinal:*

*A contrasinal é válida.*

*A contrasinal debe ter polo menos 8 caracteres.*

*A contrasinal debe conter polo menos unha letra minúscula.*

*A contrasinal debe conter polo menos unha letra maiúscula.*

*A contrasinal debe conter polo menos unha letra minúscula.*

*False*


```python
# Escribe aquí o teu código
def validar_contrasinal(contrasinal):
    if len(contrasinal) < 8:
        print("A contrasinal debe ter polo menos 8 caracteres.")
        return False
    elif not any(c.islower() for c in contrasinal):
        print("A contrasinal debe conter polo menos unha letra minúscula.")
        return False
    elif not any(c.isupper() for c in contrasinal):
        print("A contrasinal debe conter polo menos unha letra maiúscula.")
        return False
    elif not any(c.isdigit() for c in contrasinal):
        print("A contrasinal debe conter polo menos un número.")
        return False
    else:
        print("A contrasinal é válida.")
        return True


print("*Proba de validar_contrasinal:*")
resultado = validar_contrasinal("Proba123")
print(resultado)

resultado2 = validar_contrasinal("abc")
print(resultado2)
```

## Notas e Consellos sobre Funcións

Ao definir funcións en Python, seguir certas boas prácticas fará o teu código máis lexible, fácil de manter e menos propenso a erros. Aquí tes algúns consellos:

*   **Nomes descritivos:** Usa nomes claros e significativos para as túas funcións e variables. Isto axuda a entender o propósito do código sen ter que lelo en detalle. Por exemplo, `calcular_media` é mellor que `cm`.
*   **Docstrings:** Inclúe sempre un "docstring" (cadea de documentación) xusto despois da liña `def` da función. Explica brevemente que fai a función, os argumentos que recibe (`Args:`) e o que retorna (`Returns:`). Isto é fundamental para a documentación automática e para que outros (ou ti no futuro) poidan entender rapidamente como usar a túa función.
*   **Funcións curtas e dun único propósito:** Tenta que as túas funcións fagan unha única cousa ben feita (Principio de Responsabilidade Única). Se unha función se fai demasiado longa ou complexa, considera dividila en funcións máis pequenas e xestionables.
*   **Evitar efectos secundarios inesperados:** Unha función ideal non debería modificar variables fóra do seu ámbito de forma que sexa difícil de seguir. Sempre que sexa posible, as funcións deberían comunicarse co resto do código a través dos seus argumentos e valores de retorno.
*   **Comentarios (cando sexa necesario):** Usa comentarios para explicar partes do código que sexan particularmente complexas ou non evidentes. Non abuses deles; o código claro e ben nomeado moitas veces non necesita moitos comentarios. O ideal é que o código se explique a si mesmo.

Aplicar estes consellos axudarache a escribir código máis robusto e profesional.
