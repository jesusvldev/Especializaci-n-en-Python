## Exercicios sobre `if`

1.  Escribe un programa que pida ao usuario un número enteiro e imprima "O número é positivo" se o número é maior que 0.

```python
# Exercicio 1
num = int(input('Ingrese un número: '))
if(num > 0):
    print('Es positivo!')
```

2.  Escribe un programa que pida dous números e imprima "O primeiro número é maior" se o primeiro número é maior que o segundo.

```python
# Exercicio 2
num1 = int(input('Ingrese el primer numero'))
num2 = int(input('Ingrese el segundo numero'))

if(num1>num2):
    print('El primer nuemro es el mayor')
```

## Exercicios sobre `if-else`

3.  Escribe un programa que pida ao usuario un número enteiro e imprima "Par" se o número é par, e "Impar" se é impar.

```python
# Exercicio 3
num1 = int(input('Ingrese el primer numero'))
if(num1%2 != 0):
  print('impar')
else:
  print('par')
```

4.  Escribe un programa que pida a idade dunha persoa e imprima "É maior de idade" se a idade é 18 ou máis, e "Non é maior de idade" en caso contrario.

```python
# Exercicio 4
edad = int(input('Ingresa tu edad:'))
if(edad >= 18):
  print('maior')
else:
  print('menor')
```

## Exercicios sobre `if-elif-else`

5.  Escribe un programa que pida a puntuación dun exame (de 0 a 100) e imprima a cualificación segundo a seguinte escala:
    *   90-100: Sobresaliente
    *   70-89: Notable
    *   50-69: Suficiente
    *   Menos de 50: Insuficiente

```python
# Exercicio 5
nota = int(input('Ingresa tu nota:'))
if(nota >= 90 and nota <=100):
  print('sobresaliente')
elif(nota >= 70 and nota <=89):
  print('Notable')
elif(nota >= 50 and nota <=69):
  print('sufi')
elif(nota < 50):
  print('insufi')
else:
  print('nota invalida')
```
